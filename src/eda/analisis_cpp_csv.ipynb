## 1. Importar Librerías

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Configurar estilos para mejor visualización
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (12, 6)
plt.rcParams['font.size'] = 10
```

## 2. Cargar y Explorar los Datos

### 2.1. Composición del columnas

```python
# Cargar el CSV
csv_path = '../../outputs/cpp_dataset.csv'
df = pd.read_csv(csv_path)

print("Forma del dataset:", df.shape)
print("\nColumnas del dataset:")
print(df.columns.tolist())
print("\nTipos de datos:")
print(df.dtypes)
print("\nPrimeras filas:")
print(df.head())
```

### 2.2. Verificar que no haya valores faltantes

```python
print("\nValores faltantes por columna:")
print(df.isnull().sum())
```

## 3. Comparar Códigos Generados por Humanos vs IA

Análisis de la distribución entre códigos generados por humanos e inteligencia artificial.

```python
target_counts = df['target'].value_counts()
print(target_counts)

# Porcentajes de cada origen
target_pct = df['target'].value_counts(normalize=True) * 100
print("\nPorcentajes:")
print(target_pct.round(2))
```
